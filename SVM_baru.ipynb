```
import vega_datasets
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
from sklearn import svm
```

```
iris = vega_datasets.data.iris()
iris = iris.drop(iris[iris['species'] == 'virginica'].index)

sn.pairplot(iris,hue='species')
```

```
clf = svm.SVC(kernel='linear')
X = iris[['sepalWidth','sepalLength']]
y = iris['species']
clf.fit(X,y)
```

```
colors = {'setosa' : 'b', 'versicolor': 'r'}
plt.scatter(iris['sepalWidth'],iris['sepalLength'],c=[colors[r] for r in iris['species']])

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx = np.linspace(xlim[0],xlim[1],30)
yy = np.linspace(ylim[0],ylim[1],30)
XX,YY = np.meshgrid(xx,yy)
xy = np.vstack([XX.ravel(),YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

ax.contour(XX,YY,Z, colors='k', levels = [-1,0,1], alpha=0.5, linestyles = ['--','-','--'])
```
